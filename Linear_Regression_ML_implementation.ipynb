```python
cols = ['CRIM', 'ZN', 'INDUS', 'CHAS', 'NOX', 'RM', 'AGE', 'DIS', 'RAD', 'TAX',
       'PTRATIO', 'B', 'LSTAT']
```

```python
def regplot(data,x,y):
  plt.figure(figsize=(8,6))
  sns.regplot(data=data,x=x,y=y)
  plt.show()
```

```python
for col in cols:
  regplot(df,col,df['Price'])
```

```python
# Independent and Dependent features

X = df.iloc[:,0:13]
y = df.iloc[:,13:14]
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.30, random_state=42)
```

```python
# Standardize the dataset
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
```

```python
X_train = scaler.fit_transform(X_train)
```

```python
X_test = scaler.transform(X_test)
```

```python
import pickle
pickle.dump(scaler, open('scaling.pkl','wb'))
```

### Model Training

```python
from sklearn.linear_model import LinearRegression

reg = LinearRegression()
```

```python
reg.fit(X_train, y_train)
```

```python
print(reg.coef_)
```

```python
print(reg.intercept_)
```

```python
## On which parameters the model has been trained
reg.get_params()
```

```python
### Prediction with Test Data

reg_pred = reg.predict(X_test)
```

```python
# Plot a scatter plot for a prediction

plt.scatter(y_test, reg_pred)
plt.show()                   # when the plotting is linear it means your model is performed well
```

```python
## Plot residuals

sns.displot(y_test-reg_pred, kind="kde")
plt.show()
```

```python
## Scatter plot with respect to prediction and residuals

plt.scatter(reg_pred, y_test-reg_pred)
plt.show()                     # you can see uniform distibution is there with respect to test_data & residuals(y_test-predictions)
```

```python
from sklearn.metrics import mean_squared_error
from sklearn.metrics import mean_absolute_error

print(mean_squared_error(y_test,reg_pred))
print(mean_squared_error(y_test,reg_pred, squared=False))
print(mean_absolute_error(y_test,reg_pred))
```

### **R-squared & Adjusted R-Squared**

```python
from sklearn.metrics import r2_score

score = r2_score(y_test, reg_pred)
print(score)
```

###### Adjusted R2 = 1 - [(1-R2)*(n-1) / (n-k-1)]

```python
# There is no such library which could perform Adjusted R_squared so we directly implement formula

print(1-(1-score)*((len(X_test)-1)/(len(X_test)-len(X_test[0])-1)))
```
